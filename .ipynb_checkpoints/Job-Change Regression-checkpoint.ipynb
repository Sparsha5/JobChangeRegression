## Column Description

enrollee_id : Unique ID for candidate

city: City code

experience: Candidate total experience in years

company_size: No of employees in current employer's company

company_type : Type of current employer

lastnewjob: Difference in years between previous job and current job

training_hours: training hours completed

target: 0 – Not looking for job change, 1 – Looking for a job change

## Data Pre-processing

```python
df.head()
```

```python
## Dropping id columns : enrollee_id
df_wo_ids = df.drop(['enrollee_id'],axis=1)
df_wo_ids.head()
```

```python
df_wo_ids.dtypes
```

```python
#One hot encoding using label Encoder
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
df_wo_ids.loc[:,['city','gender','relevent_experience','enrolled_university',
             'education_level','major_discipline','experience','company_size','company_type','last_new_job']] = df_wo_ids.loc[:,['city','gender','relevent_experience','enrolled_university',
             'education_level','major_discipline','experience','company_size','company_type','last_new_job']].apply(enc.fit_transform)
df_wo_ids.head()
```

```python
df_wo_ids.isnull().mean()
```

```python
## Feature Selection based on Correlation
corr_mat = df_wo_ids.corr()
top_corr_features = corr_mat.index
plt.figure(figsize=(16,16))

#plot the heat map
heat_map =sns.heatmap(df_wo_ids[top_corr_features].corr(),annot=True,cmap="RdYlGn")
plt.show()

#saving the figure as png file
heat_map.figure.savefig('corr_mat.png')
```

```python
## dataset is imbalanced.. need to perform smote.
df_wo_ids['target'].value_counts()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()

cols_to_scale = ['city','training_hours','relevent_experience','enrolled_university',
             'education_level','major_discipline','experience','company_size','company_type','last_new_job']
df_wo_ids[cols_to_scale] = standardScaler.fit_transform(df_wo_ids[cols_to_scale])
df_wo_ids.head()
```

## Data upsampling using smote

```python
X = df_wo_ids.drop(['target'],axis=1)
y = df_wo_ids[['target']]
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train ,y_test = train_test_split(X,y,test_size=0.3,random_state=104) 
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=104)
X_train_smote, y_train_smote = smote.fit_resample(X_train,y_train)
```

```python
X_train_smote.head()
```

```python
y_train_smote['target'].value_counts()
```

```python
y_train['target'].value_counts()
```

## Model Building

## Linear Regression Approach

```python
## linear Regression with Ridge and lasso
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score

cv = KFold(n_splits=5,shuffle=True)
ridge = Ridge(alpha=1)
params = [{'alpha' : [1e-15 , 1e-10 , 1e-8 , 1e-3 , 1e-2, 1 , 5,10,15,20,30,40,50,60,70,80,90,100]}]
ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv = cv)
ridge_regressor.fit(X_train_smote,y_train_smote)
```

```python
ridge_regressor.best_params_ , ridge_regressor.best_score_
```

```python
cross_val_score(ridge_regressor,X_train_smote,y_train_smote,scoring='neg_mean_squared_error' , cv=5).mean()
```

```python
lasso = Lasso(alpha=1)
params = [{'alpha' : [1e-15 , 1e-10 , 1e-8 , 1e-3 , 1e-2, 1 , 5,10,15,20,30,40,50,60,70,80,90,100]}]
lasso_regressor = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv = cv)
lasso_regressor.fit(X_train_smote,y_train_smote)
```

```python
lasso_regressor.best_params_ , lasso_regressor.best_score_
```

```python
cross_val_score(lasso_regressor,X_train_smote,y_train_smote,scoring='neg_mean_squared_error' , cv=5).mean()
```

```python
## LinearRegressor's After Selecting Best params, going with Lasso

lasso_linreg = Lasso(alpha=1e-15)
lasso_linreg.fit(X_train_smote,y_train_smote)
```

```python
print (lasso_linreg.intercept_)
```

##  Decision  Tree/ RandomForest approach

```python
## Use gridsearch to get best params.
from sklearn.tree import DecisionTreeRegressor

dtree = DecisionTreeRegressor()
param_dist = {'max_depth': [3, 5, 6, 7], 'min_samples_split': [140,280,420,560,700] }
tree_grid = GridSearchCV(dtree, cv = cv, param_grid = param_dist, n_jobs = 3)
tree_grid.fit(X_train_smote,y_train_smote) 
print('Best Parameters using grid search: \n', tree_grid.best_params_)
```

```python
## Decision Tree Regression
dtree_reg = DecisionTreeRegressor(criterion='mse',max_depth=7,min_samples_split=140)
dtree_reg.fit(X_train_smote,y_train_smote)
```

```python
## Random Forest Regression params
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
param_dist = {'max_depth': [3, 5, 6, 7], 'min_samples_split': [140,280,420,560,700] }
rf_grid = GridSearchCV(rf, cv = cv, param_grid = param_dist, n_jobs = 3)
rf_grid.fit(X_train_smote,y_train_smote) 
print('Best Parameters using grid search: \n', rf_grid.best_params_)
```

```python
rf_reg = RandomForestRegressor(criterion='mse',max_depth=7,min_samples_split=140)
rf_reg.fit(X_train_smote,y_train_smote)
```

## Evaluating model

```python
## predicting test values using different models

y_pred_linear = lasso_linreg.predict(X_test)
y_pred_dtree = dtree_reg.predict(X_test)
y_pred_rf = rf_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2_linear = r2_score(y_pred_linear,y_test)
r2_dtree = r2_score(y_pred_dtree,y_test)
r2_rf = r2_score(y_pred_rf,y_test)
```

```python
mse_linear = mean_squared_error(y_pred_linear,y_test)
mse_dtree = mean_squared_error(y_pred_dtree,y_test)
mse_rf = mean_squared_error(y_pred_rf,y_test)
```

```python
print ("Linear -- r2_score : ",r2_linear, "mse : ",mse_linear)
print ("Decision tree -- r2_score : ",r2_dtree, "mse : ",mse_dtree)
print ("Random Forest -- r2_score : ",r2_rf, "mse : ",mse_rf)
```
